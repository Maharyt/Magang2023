```python
quantiles = RFM.quantile(q=[0, 0.25, 0.5, 0.75])
quantiles
```

```python
quantiles = quantiles.to_dict()
quantiles
```

```python
RFMSegmentation = RFM
```

```python
#Segmentasi RFM Method
def RClass(x,p,d):
    if x <= d[p][0]:
        return 1
    elif x <= d[p][0.25]:
        return 2 
    elif x <= d[p][0.50]:
        return 3
    elif x <= d[p][0.75]:
        return 4
    else:
        return 5
    RClass()
```

```python
#Segmentasi RFM Method
def FMClass(x,p,d):
    if x <= d[p][0]:
        return 5
    elif x <= d[p][0.25]:
        return 4
    elif x <=d[p][0.50]:
        return 3
    elif x <= d[p][0.75]:
        return 2
    else:
        return 1
    FMClass()
```

```python
RFMSegmentation['R_Quartile'] = RFMSegmentation['recency'].apply(RClass, args=('recency', quantiles,))
RFMSegmentation['F_Quartile'] = RFMSegmentation['frequency'].apply(FMClass, args=('recency', quantiles,))
RFMSegmentation['M_Quartile'] = RFMSegmentation['monetary_value'].apply(FMClass, args=('monetary_value', quantiles,))
```

```python
RFMSegmentation['RFMClass'] = RFMSegmentation.R_Quartile.map(str) \
                            + RFMSegmentation.F_Quartile.map(str) \
                            + RFMSegmentation.M_Quartile.map(str)
```

```python
# Calculate RFM_Score
RFMSegmentation['RFM_Score'] = RFMSegmentation[['R_Quartile','F_Quartile','M_Quartile']].sum(axis=1)
```

```python
# Define rfm_level function
def rfm_level(df):
    if df['RFM_Score'] >= 11:
        return "Can't Loose Them"
    elif ((df['RFM_Score'] >= 9) and (df['RFM_Score'] <= 10)):
        return 'Champions'
    elif ((df['RFM_Score'] >= 8) and (df['RFM_Score'] <= 9)):
        return 'Loyal'
    elif ((df['RFM_Score'] >= 7) and (df['RFM_Score'] <= 8)):
        return 'Potential'
    elif ((df['RFM_Score'] >= 5) and (df['RFM_Score'] <= 6)):
        return 'Promising'
    else:
        return 'Require Activation'
```

```python
# Create a new variable RFM_Level
RFMSegmentation['RFM_Level'] = RFMSegmentation.apply(rfm_level, axis=1)
```

```python
# Find K best with Elbow Method 
## Modeling with KMeans
inertia_list = []
```

```python
for i in range(1,100):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(rfm_model[['recency','frequency','monetary_value']])
    inertia_list.append(kmeans.inertia_)
```

```python
# Show graph elbow method
plt.figure(figsize=(15,10))
plt.plot(range(1,100),inertia_list,marker='o',markerfacecolor='green',linestyle='dashed')
plt.title('K value vs Inertia')
plt.xlabel('K value')
plt.ylabel('Inertia')
```
